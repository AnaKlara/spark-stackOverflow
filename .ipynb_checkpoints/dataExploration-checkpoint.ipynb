```python
db2_connection_connection = dbi.connect(db2_connection_dsn)
   
query = 'SELECT * FROM "BCZ79269"."SO_QUESTIONS"'

df = pd.read_sql_query(query, con=db2_connection_connection)
df.head()
```

```python
df['TAGS'] = df['TAGS'].str.split(',',n = 1, expand = True)
```

```python
df = df.explode('TAGS')
```

```python
df
```

```python
TAGS = df.groupby("TAGS")
TAGS.describe().head()
```

```python
TAGS.mean().sort_values(by="VIEWS",ascending=False)[["VIEWS"]].head()
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
plt.figure(figsize=(20,15))
TAGS.mean().sort_values(by="VIEWS",ascending=False)[["VIEWS"]].head(10).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("TAGS")
plt.ylabel("VIEWS")
plt.show()
```
